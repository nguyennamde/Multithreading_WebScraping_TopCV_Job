```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from time import sleep
from bs4 import BeautifulSoup
import re
import requests
import pandas as pd
import threading
import sqlalchemy
import mysql.connector
```

```python
def getURL(page_source):
    list_html_job = page_source.find_all("a", class_=False, id=False, target="_blank")
    list_url = []
    for html_job in list_html_job:
        url = html_job.get("href")
        if url not in list_url and (re.search("/viec-lam/",url)):
            list_url.append(url)
    return list_url 
```

```python
class ScrapeThreadURL(threading.Thread):
    def __init__(self, url):
        threading.Thread.__init__(self)
        self.url = url
    def run(self):
        driver = webdriver.Chrome()
        driver.get(self.url)
        sleep(2)
        page_source = BeautifulSoup(driver.page_source, "html.parser")
        url_one_page = getURL(page_source)
        url_all_page.extend(url_one_page)
        driver.close()  
        
global url_all_page
num_page = 40
url_all_page = []   
urls = [f"https://www.topcv.vn/viec-lam-it?page={page}" for page in range(1, num_page + 1)]
start = 0
end = 10
while start < end:
    threads = []
    for url in urls[start:end]:
        print(url)
        t = ScrapeThreadURL(url)
        t.start()
        threads.append(t)
    for t in threads:
        t.join()
    start=end
    
    if len(urls) - end < 10:
        end += len(urls) - end
    else:
        end+=10
```

```python
def getData(url, driver):
    driver.get(url)
    soup = BeautifulSoup(driver.page_source, "html.parser")

    container1 = soup.find("div", class_="job-detail__body-left")
    container2 = soup.find("div", class_="job-detail__body-right")
    if container1.find("a", class_="text-highlight", target="_blank") is None:
        job_title = container1.find("h1").get_text().strip()
    else:
        job_title = container1.find("a").get_text().strip()

    info = container1.find_all("div", class_="job-detail__info--section-content-value")
    info = list(map(lambda x: x.get_text().strip(), info))
    salary, address, experience = info
    company = container2.find("a", class_="name").get_text().strip()

    due_date = container1.find("div", class_="job-detail__info--deadline").get_text().strip()
    due_date = re.findall("[0-9].+", due_date)[0]

    job_description = container1.find("div", class_="job-description__item--content")
    list_description = []
    if job_description.find("li") is None:
        for description in job_description.find_all("p"):
            list_description.append(description.get_text().strip())
    else:
        for description in job_description.find_all("li"):
            list_description.append(description.get_text().strip())
    jd = " ".join(list_description)

    job_title_dict.append(job_title)
    company_dict.append(company)
    salary_dict.append(salary)
    address_dict.append(address)
    experience_dict.append(experience)
    due_date_dict.append(due_date)
    jd_dict.append(jd)
```

```python
class ScrapeThreadJob(threading.Thread):
    def __init__(self, url):
        threading.Thread.__init__(self)
        self.url = url
  
    def run(self):
        driver = webdriver.Chrome()
        getData(self.url, driver)
        driver.close()
```

```python
global job_title_dict, company_dict, salary_dict, address_dict, experience_dict, due_date_dict, jd_dict
job_title_dict = []
company_dict = []
salary_dict = []
address_dict = []
experience_dict = []
due_date_dict = []
jd_dict = []
start = 0
end = 20
        
while start < end:
    threads = []
    for url in url_all_page[start:end]:
        print(url)
        t = ScrapeThreadJob(url)
        t.start()
        threads.append(t)
    for t in threads:
        t.join()
    start=end
    if len(url_all_page) - end < 20:
        end += len(url_all_page) - end
    else:
        end+=20
list_job = {
    "job_title" : job_title_dict,
    "company" : company_dict,
    "salary" : salary_dict,
    "address" : address_dict,
    "experience" : experience_dict,
    "due_date" : due_date_dict,
    "jd" : jd_dict
}
```

```python
columns = ["job_title", "company", "salary", "address", "experience", "due_date", "jd"]
df = pd.DataFrame(list_job, columns=columns)
```

```python
df.to_csv("job_topcv.csv", index=False)
```
